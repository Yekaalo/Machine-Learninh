(a) Briefly explain how k-fold cross-validation is implemented.

> **k-fold cross validation works by first taking the data, shuffling it, then spliting the data into k-folds. Then we will hold the first fold (this will be our validation set), and use the remaining data folds to fit the model. Then we will test the model using the withheld fold. Then we repeat the same steps, but we hold the second fold as the validation set. We will repeat this k-times, until every observation has been in the validation set once. Then we find the MSE by averaging the estimates across all the k runs.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import scale
import sklearn.linear_model as skl_lm
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
import statsmodels.formula.api as smf

from sklearn import model_selection
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV
from sklearn.model_selection import train_test_split, KFold, cross_val_score
```

```python
#(a) Replace any text variables with numeric dummies

college_d = pd.get_dummies(college, drop_first=True)
college_d.head()
```

```python
#(b) Construct response y (Apps) and predictors X (the rest of variables).

y = college_d["Apps"]
X = college_d.drop("Apps", axis = 1)
```

```python
y.head()
```

```python
X.head()
```

```python
X.shape
```

```python
y.shape
```

```python
#replacing Yes/No values
X = X.replace(to_replace = ['Yes','No'],value = ['1','0'])
X
```

```python
#adding all second-order terms to our predictors
X1 = PolynomialFeatures(include_bias=False).fit_transform(X)
X1.shape
```

```python
#Split the data set into a training set and a test set (second order predictors)
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y)
```

```python
##Split the data set into a training set and a test set (original predictors)

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#standardize all explanatory variables
X1_scaled = scale(X1)
X1s_train, X1s_test, y_train, y_test = train_test_split(X1_scaled, y)
```

```python
X1s_train.shape
```

```python
#Check scaled mean = 0
X1_scaled.mean()
```

```python
#Check scaled SD = 1

X1_scaled.std()
```

```python
#Fit a linear model using least squares on the training set

reg = skl_lm.LinearRegression(fit_intercept = True)
reg.fit(X1s_train,y_train)

y_pred = reg.predict(X1s_test)

print(reg.intercept_)
print(reg.coef_)
```

```python
print(mean_squared_error(y_test, y_pred))
print(reg.intercept_)
print(reg.coef_)
errors_LR = abs(y_pred - y_test)
# Calculate mean absolute percentage error (MAPE)
mape_LR = 100 * (errors_LR / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape_LR)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
#Ridge regression

alphas = 10**np.linspace(5,-2,100)*0.5

ridge = Ridge()
coefs = []

for a in alphas:
    ridge.set_params(alpha=a, fit_intercept = True)
    ridge.fit(scale(X), y)
    coefs.append(ridge.coef_)

ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
#ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('Standardized Coefficients')
plt.title('Ridge coefficients as a function of the regularization')
plt.legend(X.columns[:4])
```

The above plot shows that the Ridge coefficients get larger when we decrease alpha.



```python
#cross validation for alphas
alphas = 10**np.linspace(-5, 3,100)

ridge = Ridge()
cv_r = []

for a in alphas:
    ridge.set_params(alpha=a, fit_intercept = True)
    ridge.fit(scale(X1_train), y_train)
    cv_r.append(cross_val_score(ridge, scale(X_train), y_train, cv=5, scoring='neg_mean_squared_error').mean())

```

```python
#cross validation ridge
alphas = 10**np.linspace (-5,3,100)

ridgecv = RidgeCV(alphas=alphas, fit_intercept=True) #scoring='neg_mean_squared_error')
ridgecv.fit(scale(X1_train), y_train)

mean_squared_error(y_test, ridgecv.predict(scale(X1_test)))
```

```python
#Plot cross-validated MSE as a function of λ i.e. alpha
(-pd.DataFrame(cv_r, index=alphas)).plot()
```

```python
#MSE test

mean_squared_error(y_test, ridge.predict(scale(X1_test)))
```

```python
#lasso regression

lasso = Lasso()
coefs = []
alphas = 10**np.linspace(5,-2,100)*0.5


for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(scale(X1_train), y1_train)
    coefs.append(lasso.coef_)


ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
#ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Lasso coefficients as a function of the regularization')
plt.legend(X.columns[:4])
```

```python
#Find optimal penalty based on cross-validation
alpha_l = np.linspace(0.001, 2, num=1000)

lassocv = LassoCV(alphas=alpha_l, fit_intercept = True, cv=5)
lassocv.fit(scale(X_train), y_train)

mean_squared_error(y_test, lassocv.predict(scale(X_test)))
```

```python
#kkk

rng = np.random.RandomState(42)
x = 10 * rng.rand(200)

def model(x, sigma=0.3):
    fast_oscillation = np.sin(5 * x)
    slow_oscillation = np.sin(0.5 * x)
    noise = sigma * rng.randn(len(x))

    return slow_oscillation + fast_oscillation + noise

y = model(x)
plt.errorbar(x, y, 0.3, fmt='o');
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(200)
forest.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = forest.predict(xfit[:, None])
ytrue = model(xfit, sigma=0)

plt.errorbar(x, y, 0.3, fmt='o', alpha=0.5)
plt.plot(xfit, yfit, '-r');
plt.plot(xfit, ytrue, '-k', alpha=0.5);
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split


# Create the hyperparameter grid
l1_space = np.linspace(0, 1, 20)
param_grid = {'l1_ratio': l1_space}

# Instantiate the ElasticNet regressor: elastic_net
elastic_net = ElasticNet()

# Setup the GridSearchCV object: gm_cv
gm_cv = GridSearchCV(elastic_net, param_grid, cv=5)

# Fit it to the training data
gm_cv.fit(X_train, y_train)

# Predict on the test set and compute metrics
y_pred = gm_cv.predict(X_test)
r2 = gm_cv.score(X_test, y_test)
```
